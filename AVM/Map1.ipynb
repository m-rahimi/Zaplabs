```python
path = ""
print("Read train file to obtain some statistic about the features and target")
df = pd.read_csv(path + 'alameda_clean.csv')
print("the shape of the data frame:", df.shape)
```

```python
df2 = df.sample(2000)
```

```python
df2.columns
```

```python
df2[["LATITUDE", "LONGITUDE", "INACT_SELLING_PRICE"]].describe()
```

```python
import branca.colormap as cm
linear = cm.LinearColormap(
    ['green', 'yellow', 'red'],
    vmin=6e5, vmax=10e5
)

linear
```

```python
import folium
map1 = folium.Map(location=[df.LATITUDE.mean(), df.LONGITUDE.mean()], zoom_start=10.4, min_zoom=10.5)
```

```python
for ii, row in df2.iterrows():
    folium.CircleMarker(location=[row["LATITUDE"], row["LONGITUDE"]], radius=.4, 
                        color=linear(row["INACT_SELLING_PRICE"])).add_to(map1)
```
